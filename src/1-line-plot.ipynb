```python

# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = x**2  # y values as the square of x

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2", color="blue", linewidth=2)

# Customize the plot
plt.title("Line Plot of y = x^2", fontsize=16)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```
