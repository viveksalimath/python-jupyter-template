```python

# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart in the first subplot
axs[0, 0].plot(x, y, color='blue', label='y = x^2')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("x values")
axs[0, 0].set_ylabel("y values")
axs[0, 0].legend()

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', s=100)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x values")
axs[1, 0].set_ylabel("y values")

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
